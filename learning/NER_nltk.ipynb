```python
import nltk
from nltk.corpus import state_union
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
data = state_union.raw("2006-GWBush.txt")
```

```python
sentences = [sent_tokenize(data)]
```

```python
sentences
```

```python
words = [word_tokenize(data)]
```

```python
words
```

```python
tagged = [nltk.pos_tag(word) for word in words]
```

```python
tagged[:1]
```

```python
ner_tags = [nltk.ne_chunk(tag) for tag in tagged]
```

```python
ner_tags[0]
```

```python
entities = list()
```
